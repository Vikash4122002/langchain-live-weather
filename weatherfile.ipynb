```python
from dotenv import load_dotenv
load_dotenv()

from langchain_groq import ChatGroq
from langchain_community.tools import DuckDuckGoSearchRun
from langchain_core.prompts import ChatPromptTemplate
```

```python
llm = ChatGroq(
    model="llama-3.1-8b-instant",
    temperature=0
)
```

```python
search = DuckDuckGoSearchRun()
```

```python
prompt = ChatPromptTemplate.from_messages([
    (
        "system",
        "You are a weather assistant. "
        "Use ONLY the web search results provided. "
        "Summarize the current weather clearly. "
        "Always include sources."
    ),
    ("human", "{question}\n\nWeb search results:\n{search_results}")
])
```

```python
def get_kolkata_weather():
    search_results = search.run(
        "current live weather in Kolkata India today"
    )
    
    response = llm.invoke(
        prompt.format_messages(
            question="What is the current live weather in Kolkata, India?",
            search_results=search_results
        )
    )
    
    return response.content
```

```python
print("===== KOLKATA LIVE WEATHER =====\n")
print(get_kolkata_weather())
```
